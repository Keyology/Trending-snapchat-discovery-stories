```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data set 
csv_file_name = 'best-snapchat-discover-stories-dataset.csv'
data_set = pd.read_csv(csv_file_name)
```

```python
# remove none university from dataset 
data_set = data_set[data_set.segment_type == 'University']
# remove people have not wantched any of the content in dataset
data_set = data_set[data_set.answer != 'Never watched those / other']

```

```python
# Get the most trending content by university
data_set = data_set.groupby(['segment_description', 'answer'])['count'].sum().sort_values(ascending=False).reset_index().drop_duplicates('segment_description', keep='first')
```

```python
data_set.replace(0, np.nan, inplace=True)
```
